```python
# It consists of only 20 values in the first page
# Now we need to do this in loop which will run for no. of pages times
# We will create new dataframe in each loop and will append it in the earlier dataframes
# such that we will ultimately get a large dataframe with 8000+ movies
```

```python
# We will create empty dataframe 
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
# &language=en-US&page={}'.format(i))
```

```python
for i in range (1,497):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=b1573681e41c377765bda65aef3b74cf&language=en-US&page={}'.format(i)) 
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True) 
```
